## 步骤 2：基础对话

定义通用对话函数，并做一个简单的连通性测试。

```python
def chat_with_llm(messages, model=MODEL, temperature=0.7, max_tokens=1024):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content


# 连通性测试
print("正在进行 API 连通性测试...\n")

try:
    test_reply = chat_with_llm(
        [{"role": "user", "content": "你好，请用一句话介绍番茄的种植要点。"}],
        temperature=0.3,
        max_tokens=100,
    )
    print(f"LLM 回复: {test_reply}")
    print("\nAPI 连接成功！")
except Exception as e:
    print(f"API 连接失败: {e}")
    print("请检查 API Key 和网络连接。")
```

## 步骤 3：Prompt Engineering 三组对比实验

比较三种不同的 Prompt 策略，观察回答质量和专业性的差异。

```python
def benchmark_prompt(system_prompt, user_prompt, label=""):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt},
    ]
    start = time.time()
    reply = chat_with_llm(messages, temperature=0.3)
    elapsed = time.time() - start
    print(f"\n{'=' * 60}")
    print(f"【{label}】")
    print(f"耗时: {elapsed:.1f}s")
    print(f"回答:\n{reply}")
    return reply


# 实验 1：基础 Prompt（无角色设定，无格式要求）
print("\n" + "#" * 60)
print("  实验 1：基础 Prompt")
print("#" * 60)

benchmark_prompt(
    system_prompt="",
    user_prompt="番茄叶子发黄是什么原因？",
    label="实验1：基础 Prompt",
)
```

```python
# 实验 2：角色设定 + 结构化输出要求
print("\n" + "#" * 60)
print("  实验 2：角色设定 + 结构化输出")
print("#" * 60)

benchmark_prompt(
    system_prompt="你是一位有20年经验的农业植保专家。请用专业但易懂的语言回答农民的病害问题。",
    user_prompt=(
        "番茄叶子发黄是什么原因？"
        "请从以下方面分析："
        "1. 可能的病害类型\n"
        "2. 非病害原因（营养、水分等）\n"
        "3. 如何区分\n"
        "4. 防治建议"
    ),
    label="实验2：角色设定+结构化",
)
```

```python
# 实验 3：Few-shot + Chain of Thought（逐步推理）
print("\n" + "#" * 60)
print("  实验 3：Few-shot + 逐步推理")
print("#" * 60)

benchmark_prompt(
    system_prompt="你是一位农业植保专家。请按以下步骤分析：先观察症状特征，再对比已知病害，最后给出诊断和建议。",
    user_prompt="""请参考以下诊断示例，对新的症状进行分析：

示例1：
- 症状：番茄叶片出现同心轮纹状褐色病斑
- 诊断：早疫病（Alternaria solani）。轮纹状病斑是典型特征。
- 建议：移除病叶，喷施代森锰锌或苯醚甲环唑。

示例2：
- 症状：番茄果实出现水渍状褐色凹陷斑，高湿环境下有白色霉层
- 诊断：晚疫病（Phytophthora infestans）。水渍状斑和白色霉层是关键特征。
- 建议：控制湿度，喷施烯酰吗啉或霜脲氰。

新症状：番茄叶片边缘出现 V 形黄褐色坏死斑，病健交界明显，多从下部叶片开始。
请逐步分析。""",
    label="实验3：Few-shot+逐步推理",
)
```

## 步骤 4：多轮对话

模拟一次完整的农业咨询对话，展示 LLM 如何基于上下文进行连续交流。

```python
print("=" * 60)
print("  多轮对话：模拟农业咨询")
print("=" * 60)

messages = [
    {
        "role": "system",
        "content": "你是农业植保专家，专门帮助农民诊断和防治作物病害。回答要专业、实用、有依据。",
    }
]


def multi_turn_chat(user_input):
    messages.append({"role": "user", "content": user_input})
    reply = chat_with_llm(messages, temperature=0.5)
    messages.append({"role": "assistant", "content": reply})
    print(f"\n农民: {user_input}")
    print(f"\n专家: {reply}")
    return reply


# 第一轮：描述症状
multi_turn_chat("我家的番茄最近叶子开始出现褐色斑点，集中在下部叶片，是怎么回事？")

print("\n" + "-" * 40)

# 第二轮：询问用药
multi_turn_chat("用什么药比较好？我现在手头有代森锰锌。")

print("\n" + "-" * 40)

# 第三轮：询问使用方法
multi_turn_chat("打药的频率应该是多少？")
```

## 步骤 5：可视化对比

用柱状图直观展示三种 Prompt 策略的回答质量差异。

```python
# 可视化对比三种 Prompt 策略
experiments = ["基础 Prompt", "角色+结构化", "Few-shot+推理"]
quality_scores = [3, 7, 9]  # 主观评分 1-10

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(
    experiments,
    quality_scores,
    color=["#4e79a7", "#f28e2b", "#e15759"],
    edgecolor="black",
)

for bar, score in zip(bars, quality_scores):
    ax.text(
        score + 0.2,
        bar.get_y() + bar.get_height() / 2,
        f"{score}/10",
        va="center",
        fontsize=12,
        fontweight="bold",
    )

ax.set_xlabel("回答质量评分（主观）", fontsize=12)
ax.set_title("Prompt Engineering 效果对比", fontsize=14, fontweight="bold")
ax.set_xlim(0, 11)
ax.grid(True, alpha=0.3, axis="x")
plt.tight_layout()
plt.savefig("task_a_prompt_comparison.png", dpi=150, bbox_inches="tight")
plt.show()

print("对比图已保存到 task_a_prompt_comparison.png")
```

### 实验结果分析

### 思考题

**1. 为什么角色设定能显著提升 LLM 的回答质量？**

角色设定相当于给模型指定了一个特定的知识领域和表达风格。LLM 在训练时学习了海量不同角色和领域的文本，角色设定可以激活模型在该领域相关的知识和表达模式，使回答更加专业和一致。

**2. Few-shot learning 中，示例的数量和质量哪个更重要？**

质量比数量更重要。2-3 个高质量、有代表性的示例就能显著改善输出。关键是示例要准确、结构清晰，能够展示期望的输出格式和推理过程。

**3. 在农业场景下，结构化输出为什么特别重要？**

农业病害诊断需要条理清晰的分析（症状→病因→防治），结构化输出让农民一目了然，便于实际操作。混乱的信息在实际生产中可能导致误判和损失。
